## Google AI API (Beispiel für eine REST-API mit API-KEY)
1. Man benötigt einen API-Key den man nach der Anmeldung erzeugen kann:
https://aistudio.google.com/apikey

2. Es muss ein HTTP-Request mit folgenden Bestandteilen erzeugt werden.
- url (die Ziel-Url)
- header (Meta-Informationen die benötigt werden u.a. API-Key)
- methode (HTTP-Request-Methode: POST)
- daten (die Daten die versendet werden)

Wir verwenden dazu wieder das requests-Modul. Aber statt `.get()` müssen wir diesmal die Methode `.post()` verwenden.
Dieser Methode werden `url`, `header` und `daten` übergeben.


Das HTTP-Request-Pattern verät uns welche Informationen wir übertragen müssen:

```
curl "https://generativelanguage.googleapis.com/v1beta/models/gemini-2.0-flash:generateContent" \
  -H 'Content-Type: application/json' \
  -H 'X-goog-api-key: GEMINI_API_KEY' \
  -X POST \
  -d '{
    "contents": [
      {
        "parts": [
          {
            "text": "Explain how AI works in a few words"
          }
        ]
      }
    ]
  }'
```

````python
#3
#vgl.:
#https://ai.google.dev/gemini-api/docs/quickstart?lang=python&hl=de#rest


import os


GEMINI_API_KEY = os.environ["GEMINI_API_KEY"]
#GEMINI_API_KEY = "" # Alternativ direkt eintragen


url = r"https://generativelanguage.googleapis.com/v1beta/models/gemini-2.5-flash:generateContent"

headers = {'Content-Type':'application/json',"x-goog-api-key": GEMINI_API_KEY} 
# Header-Informationen müssen für request als dictionary übergeben werden
#print(headers)

data = {
    "contents": [
      {
        "parts": [
          {
            "text": "Explain how AI works in a few words"
          }
        ]
      }
    ]
  }

import requests, json

def jsonPrint(obj):
    text = json.dumps(obj,sort_keys=True, indent=4)
    print(text)


response = requests.post(url, headers=headers, json=data)

if response.status_code == 200:
    result = json.loads(response.text)
    print(result)
else:
    print("Fehler beim Laden der Resource")
#print(response)
#print(result)
jsonPrint(result)

````

````python
#5
from google import genai

# The client gets the API key from the environment variable `GEMINI_API_KEY`.
if not os.environ["GEMINI_API_KEY"]:
    os.environ["GEMINI_API_KEY"] = GEMINI_API_KEY


client = genai.Client()

response = client.models.generate_content(
    model="gemini-2.5-flash", contents="Erkläre das Gaussche Lösungs-Verfahren für Lineare Gleichungs-Systeme"
)

#Ergebnis 
from IPython.display import display_markdown
display_markdown(response.text,raw=True) 
````
