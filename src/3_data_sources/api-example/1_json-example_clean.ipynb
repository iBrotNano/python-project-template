### JSON-Format

- JSON ist eine "Notation" für Daten um diese zu übertragen oder zu speichern
- eine alternative zu XML, welche effizient zu lesen und zu schreiben ist
- JSON ist ideal für den schnellen Datenaustausch. XML hingegen bleibt dort überlegen, wo Dokumentenstrukturen, strenge Validierung oder komplexe Datentransformationen erforderlich sind
