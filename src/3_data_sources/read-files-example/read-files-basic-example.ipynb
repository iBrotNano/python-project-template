## Einlesen und Schreiben von Daten


#### In eine Datei schreiben

```python
datei = open('test.txt','w') # 'r' = read, 'a'= append , 'w'= write/overwrite
datei.write("Hallo\n")
datei.write("Python!")
datei.write("\n")
datei.close()
```

- die Datei wird relativ zum Projekt-Verzeichnis erzeugt 
- Es gibt _relative_ und _absolute_ Pfade
    - __absolute__ Pfade beginnen immer "/" und werden i.d.r. ab dem Wurzel-Verzeichnis angegeben
    - __relativen__-Pfade beginnen NICHT mit "/" und werden relativ zu dem Verzeichnis defininiert 
    von dem aus der Pyhton-Interpreter gestartet wurde

#### Eine Datei einlesen

```python
datei = open('test.txt','r')
txt = datei.read()
datei.close()
print(txt)
```

#### `with` verwenden
- Vorteil: es muss kein `.close()`verwendet werden

```python
with open('test.txt','r') as datei:
    print(datei.read())


# write analog
```

#### Datei Zeilenweise lesen

```python
with open('test.txt','r') as datei:
    for zeile in datei:
        print(zeile,end="")


datei = open('test.txt','r')
for zeile in datei:
    print(zeile,end="")
datei.close()

```

#### Telefonnummern aus einer Datei lesen

```python
with open('example.txt','r') as datei:
    example_txt = datei.read()

import re

numbers = re.findall(r'[0-9\s()-]{6,20}',example_txt)
numbers = [x.strip() for x in numbers]
numbers
```

#### Alle Links aus einem HTML-Dokument extrahieren

```python
with open('website.html','r') as datei:
    website_html = datei.read()

import re

links = re.findall(r'href=\".+?html\"',website_html)
links = [ re.search(r'\".+?\"',x).group() for x in links]
links = [ x.strip('"') for x in links]
links
```

### `CSV`-Modul

- Optional: Installation von CSVLint-Extension (Farbliche Unterscheidung von Spalten)


```python
import csv

#CSV schreiben
data = [
    ['Name', 'Alter', 'Abteilung'],
    ['Julia', '30', 'Marketing Abteilung'],
    ['Dieter', '45', 'Logistik \'Abteilung']
    ['Dieter', '45', 'Logistik , Abteilung']
]

with open('employees.csv', 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file,quotechar="'",delimiter=',',quoting=csv.QUOTE_MINIMAL)
    #vgl. writer = csv.writer(file,quotechar="'",delimiter=',',quoting=csv.QUOTE_ALL)
    writer.writerows(data) # Alle Zeilen unmittelbar schreiben

# CSV lesen
with open('employees.csv', 'r', encoding='utf-8') as file:
    reader = csv.reader(file, delimiter=',', quotechar="'")
    for row in reader:
        print(row) 

#['Name', 'Alter', 'Abteilung']
#['Julia', '30', 'Marketing Abteilung']
#['Dieter', '45', 'Logistik ,Abteilung']

```
