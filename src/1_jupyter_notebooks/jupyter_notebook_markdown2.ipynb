```markdown
* Listen-Element 1
* Listen-Element 2
    * Listen-Element 2.1
    * Listen-Element 2.2
  
1. Listen-Element 1
2. Listen-Element 2  
    3. Listen-Element 1
```

```markdown
## LaTex Markup verwenden z.B. Mathematische Formeln:

* Eine Latex-Formel ist mit $ umschlossen  

Beispiele:  

`$a^{2x}$`, $a^{2x}$
  
`$a_{2x}$`, $a_{2x}$

`$\frac{1}{n}$`, $\frac{1}{n}$

`$\sum\nolimits_{n=0}^N$ a^{n}`, $\sum\nolimits_{n=0}^N a^{n}$

`$\sqrt{9}$`, $\sqrt{9}$
```
