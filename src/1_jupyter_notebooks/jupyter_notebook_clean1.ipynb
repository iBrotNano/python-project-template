### Jupyter-Magic-Befehle mit `%`

- Jupyter-Magic-Commands sind spezielle Funktionen, die in den IPython-Kernel integriert sind und <u>___zusätzliche Funktionalität___</u> bieten, die über das Standard-Python hinausgeht.

```python
#5 
#Externes Script ausführen

%run myscript.py
```

```python
#6
#Ausführungszeit messen 

#%timeit? # Informationen 
%timeit [x**2 for x in range(1000)] # Zeit messen
```

```python
#7
%magic # Hilfe zu Magics anzeigen
```

### Shell-Befehle
- Shell-Befehle können in Code-Zellen mittels `!` ausgeführt werden
- <u>Achtung:</u> Shell-Befehle werden immer relativ zum ausführenden Script ausgeführt
- Shell-Ausgaben können in Python verwenden, sowie Pyhton-Variablen in Shell-Befehlen

```python
#9
# Typische Shell-Befehle
#!pwd
#!echo "hallo"
#!ls
#!mkdir data
#!pip install package # Möglich aber besser mit Magic-Befehl s.u.

# Shell-Ausgaben in Python verwenden
#content = !ls
#print(content)

# Python-Variablen in Shell-Befehl verwenden
#!ls {content[2]} # 
```

- Shellbefehle werden in einer _temporären Sub-Shell_ ausgeführt, daher funktioniert beispielsweise `!cd` nur temporär für den selben Befehl
