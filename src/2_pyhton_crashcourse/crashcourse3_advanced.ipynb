
#### Iterables, Generatoren
- Listen, Tuple, Dictionaries sind _Iterables_, d.h. man kann über sie mittels einer Schleife iterieren
- Eine weitere Möglichkeit zur verwendung in Schleifen sind ___"Generatoren"___
- __Python-Generatoren__ bieten durch _Lazy Evaluation_ (verzögerte Auswertung) enorme Speicher- und Performance-Vorteile, da sie Werte einzeln bei Bedarf erzeugen, statt sie alle im Voraus im Arbeitsspeicher zu halten
- Eine Funktion wird zu einem Generator wenn Sie `yield` statt `return` verwendet. Nach dem `yield` wird der Zustand der Funktion gespeichert. Alle in den Variablen gespeicherten Werte bleiben erhalten.
- beim nächsten Aufruf eines Generators wird unmittelbar nach dem letzten `yield` fortgefahren.
- Ein Nachteil von Generatoren ist, daß man nur einmal über sie iterieren kann
- erst wenn die Generator-Funktion komplett durchlaufen ist, endet auch die Iteration der verwendenden Schleife. Dann hat die Funktion den impliziten Rückgabewert `None`.
- `return None` stoppt also den Generator




```python
#6
def generate_range(n):
    for i in range(n):
        yield(i) # beim nächsten Aufruf des Generators wird nach dem letzten yield fortgefahren bis die Funktion beendet ist.
        #yield(i*i) # Es können sogar auch mehrere yield Statements verwendet werden
        #print("test",i)
        
        # return None stoppt den Generator 
        #if i == 1:
        #    return None
    

i=0
my_generator = generate_range(3)
for k in my_generator: 
    print(f"Index:{i} ",f"Wert:{k} ")
    input() # <=
    
    i+=1


# das nächste Element wird dabei mit der next() Funktion abgerufen
     
gen = generate_range(3)
print(next(gen))
print(next(gen))


```

## Reguläre Ausdrücke
- Reguläre Ausdrücke sind ein **mächtiges Werkzeug** um Zeichenketten zu Untersuchen und zu manipulieren
- Reguläre Ausdrücke können kompliziert sein
- Python bietet das Modul "__re__" um __Reguläre-Ausdrücke__ zu verwenden 

```python
#7
# Erste Anwendung: Suchen nach Pattern in Strings mit der Funktion "search()"

import re

txt = "The rain in Spain"
x = re.search("Portugal", txt)
print(x) # None

x = re.search("ain", txt) # liefert ein re.Match object, welches dem "ersten" Match entspricht
#x ist ein Match objekt
print(type(x),x)#<class 're.Match'> <re.Match object; span=(5, 8), match='ain'>
print(x,bool(x))# lässt sich aber per bool() umwandeln



```

```python
#8
#x ist ein Match-Objekt und hat daher eigene Methoden und Eigenschaften
print(x.span())
print(x.string)
print(x.group())


#.span() returns a tuple containing the start-, and end positions of the match.
#.string returns the string passed into the function
#.group() returns the part of the string where there was a match (hier trivial, der Suchstring)
```

### Ein Paar Beispiele:

```python
#10
#Paar Beispiele
txt = "The rain in Spain"
x = re.search("^The.*Spain$", txt) #  ^begin, $ end, . Platzhalter, * = 0 oder mehr Vorkommen von "."
print(r"^The.*Spain$",bool(x)) #True
# Es empfiehlt sich Raw-Strings zu verwenden für das Pattern, sonst kommt es teilweise zu fehlern

x = re.search("^T.*Spain$", txt) 
print(r"^T.*Spain$",bool(x)) #True

x = re.search("^.*Spain$", txt)
print(bool(x))#True
x = re.search("^All.*Spain$", txt)
print(bool(x))#False
```

```python
#11
# Gültige Deutsche Telefonnummer prüfen

txt = "+49 123456"
x = re.search(r"^\+49[\s\-0-9]+$", txt) # Gültige Deutsche Telefonnummer
print("+49 123456",bool(x)) # True

#Achtung:
# Ohne Raw-String wird \ als Escape-Sequence interpretiert
# => <>:4: SyntaxWarning: invalid escape sequence '\+'
# da "\+" keine gültige Escape-Sequenz ist 

txt = "123+456"
x = re.search(r"^\+49[\s\-0-9]+$", txt)
print("123+456",bool(x))#False

txt = "0049 123456"
x = re.search(r"^\+49[\s\-0-9]+$", txt)
print("0049 123456",bool(x))#False


txt = "12345"
x = re.search(r"[0-9]{5}$", txt)
print("12345",bool(x))

txt = "1234"
x = re.search(r"[0-9]{5}$", txt)
print("1234",bool(x))


#vgl.: https://www.w3schools.com/python/python_regex.asp
# u.a.

# Achtung:
# Manche Steuerzeichen haben je nach Kontext verschiedene Bedeutungen
# z.B. in Sets => bedeutet [^abc] "NICHT" die Buchstaben a,b, oder c
# Am Anfang eines Regulären Ausdrucks wird durch ^ der Beginn des Suchstrings markiert

```

#### Funktions-Parameter übergabe mit <code>*args</code>
- mit * ("Asterisk") markierte Parameter sind flexibel. Es lassen sich beliebig viele Werte an die Funktion übergeben
- in der Funktion ist das Argument ein Tupel
- beim Aufruf der Funktion werden die Paramete wie postionale Parameter werwendet

```python
#18
# *args
def magic(*args):
    #print(type(args)) # Tuple
    for i in args:
        print(i)

magic("test",0.1, [])
print("---")
magic("test",0.1, [],5)




```

#### Funktions-Parameter übergabe mit <code>**kwargs</code> (Key-Word-Arguments)
- key-word-Arguments markiert mit **
- **varname beinhaltet eine beliebige Anzahl von Keyword-Parametern
- type(kwargs) ist ein Dictionary

```python
#19
def magic(**kwargs):
    print(type(kwargs)) # <class 'dict'>
    for i,j in kwargs.items():
        print(i,j)

magic(a=1,b=2)
print("---")
magic(a=1,b=2,c=3,d=4)
```

#### Zip ("Zipper" - Reißverschluss)
- Zwei Listen zu einer kombinieren

```python
#20
x = [0,1,2,3,4,5,6,7]
y = [2,3,4,5,6,7,8,9]

print(zip(x,y)) # zip-objekt ist ein Iterator

zipped = list(zip(x,y)) # kann in eine Liste umgewandelt werden
print(zipped)

# Was passiert bei unterschiedlich langen Listen ?
```

```python
#21
# zip ist flexibel, funktioniert auch mit Dictionaries
zipped = dict(zip(x,y))
print(zipped)# {0: 2, 1: 3, 2: 4, 3: 5, 4: 6, 5: 7, 6: 8, 7: 9}

# Der umgekehrte Weg (funktioniert aber nicht mit Dictionaries)

zipped = list(zip(x,y))
print(zipped) #[(0, 2), (1, 3), (2, 4), (3, 5), (4, 6), (5, 7), (6, 8), (7, 9)]
print(*zipped) # (0, 2) (1, 3) (2, 4) (3, 5) (4, 6) (5, 7) (6, 8) (7, 9)
m,n = zip(*zipped)
print(m) # (0, 1, 2, 3, 4, 5, 6, 7)
print(n) # (2, 3, 4, 5, 6, 7, 8, 9)
# * führt das Entpacken der Elemente durch, kann nur innerhalb einer Parameterliste verwendet werden (!)

#analoges Beispiel:
m,n = zip((1,2),(3,4),[5,6]) # Argumente müssen Iterables sein
print(m) #(1, 3, 5)
print(n) #(2, 4, 6)

```
