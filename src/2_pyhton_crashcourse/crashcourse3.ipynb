
## Buildin-Datenstrukturen
### Listen


````python
#8
# Buildin-Datenstrukturen

# Listen
mylist =[1,2.,"hello",["a","b"]]# Listen-Elemente können heterogen sein
print(mylist) # [1, 2.0, 'hello', ['a', 'b']]

# Zugriff per Indexing und Slicing
# Beim Indexing wird immer das Element zurückgegeben
# Beim Slicing, - auch wenn es nur ein Element ist, immer eine Liste

print(mylist[0]) # 
print(mylist[-1])#

print(mylist[1:2]) 
print(mylist[::-1]) # 
print(mylist[:3:]) 



# Beim Slicing wird immer eine neue Liste erzeugt (!)

new_list = mylist[:] # Kopie (!)
new_list = mylist.copy() # Kopie "explizit" mit methode .copy()

new_list[0] = "neuer wert"
print(new_list,mylist) # nur new_list wurde geändert

# sonst wird bei "mutable Datentypen" (set ,list and dict) per Referenz übergeben:
new_list2 = mylist
new_list2[0] = 0
print(new_list2,mylist) # beide Listen wurden geändert !!

# immutable Build-in Datentypen sind: int, float, bool, string, tuple



````

````python
#31

# Lambda Funktionen sind Anonyme Funktionen
# Eine Lambda-Funktion kann eine beliebige Anzahl von Argumenten annehmen, aber nur einen Ausdruck/Ergebnis haben.

#def mylambda(x,y):
#    return 2*x +y

# Das gleich könnte man mit lambda-Schreibweise so implementieren:
mylambda = lambda x,y: 2*x +y # f(x,y)= 2*x+y

print(mylambda(8,2))


def doSomethingWithFunction(f):
    return f(7)

val = doSomethingWithFunction(lambda x: 2*x)
print(val)

my_function = lambda x: x**(1/2)
val = doSomethingWithFunction(my_function) # wurzel aus 7 
print(val)
````

````python
#40
#Bsp.:

x = list() # Wir erzeugen ein Listen-Objekt
x.append(2) # per Punkt-Notation werden die Methoden aufgerufen
x.append(4) # Methoden werden immer auf ein Objekt angewendet, welches i.d.r erst erzeugt werden muss
x.append(-1)
x.sort()
print(x)

#"normale" Funktionen brauchen kein Objekt um ausgeführt zu werden

x =[4,6,3,1] 
y = sorted(x)# hier wird eine neue Liste erzeugt
print(y)

print(x) # x weiterhin unsortiert
x.sort()# wärend hier die Liste x sortiert wird
print(x)

````



## Objekte können mittel Klassen beschrieben werden, welche als Vorlage zum erzeugen der Objekten dienen
* Ein Klasse beschreibt die Methoden und Eigenschafte eines Objekts
* Als Anwender muss man nicht unbedingt den Aufbau der Klasse kennen, stattdessen dienen die Methoden als "Schnittstelle":  
Bsp.: 
    ```python
    cat1 = Cat("Kitty", 2.5) # Erzeugung des Obektes. 
    cat1.make_sound() # Methoden aufruf. 
    cat1.change_name("Simba") # Manipulation der Eigenschaft "name". 
    ```

* Klassen und Vererbung sind im Data-Science-Kontext weniger wichtig, da man eher existierende Klassen anwendet als neue zu schreiben. 




````python
#41
# Ein Objekt wird über eine Klasse definiert mit gewissen Eigenschaften und Methoden
 
class Cat:
    # Eigenschaften name und age
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # Methoden, - eine Aktion auf einem Objekt
    def info(self):
        print(f"I am a cat. My name is {self.name}. I am {self.age}")

    def make_sound(self):
        #self.info()
        print("Meow")


class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def info(self):
        print(f"I am a dog. My name is {self.name}. I am {self.age}")

    def make_sound(self):
        print("Bark")


cat1 = Cat("Kitty", 2.5)
dog1 = Dog("Fluffy", 4)

for animal in (cat1, dog1):
    animal.make_sound()
    animal.info()

````

### Vererbung

````python
#42
# Vererbung

class GermanShepard(Dog):
    
    def __init__(self):
        super().__init__("Blondi",92)
        #equivalent zu:
        #self.name = "Blondi"
        #self.age = "92"
    

    def info(self): # Overwrite
        print(f"I am a German Shepard.")

g = GermanShepard()# Objekt erzeugen

g.info() 
g.make_sound()

# Vererbung ist u.a. nützlich um Schnittstellen bereitzustellen 
# - die Komplexität kann verborgen werden, stattdessen werden nur Anwendungsbezogene Funktionen für Anwender zu Verfügung gestellt. 
# - Wiederverwendbarkeit wird gefördert
# - Vererbung wird im Kontext vom MapReduce-Algorithmus vom Framework MRJob verwendete und bildet dort die Schnittstelle für den Anwender


````
