
#### Iterables, Generatoren
- Listen, Tuple, Dictionaries sind _Iterables_, d.h. man kann über sie mittels einer Schleife iterieren
- Eine weitere Möglichkeit zur verwendung in Schleifen sind ___"Generatoren"___
- __Python-Generatoren__ bieten durch _Lazy Evaluation_ (verzögerte Auswertung) enorme Speicher- und Performance-Vorteile, da sie Werte einzeln bei Bedarf erzeugen, statt sie alle im Voraus im Arbeitsspeicher zu halten
- Eine Funktion wird zu einem Generator wenn Sie `yield` statt `return` verwendet. Nach dem `yield` wird der Zustand der Funktion gespeichert. Alle in den Variablen gespeicherten Werte bleiben erhalten.
- beim nächsten Aufruf eines Generators wird unmittelbar nach dem letzten `yield` fortgefahren.
- Ein Nachteil von Generatoren ist, daß man nur einmal über sie iterieren kann
- erst wenn die Generator-Funktion komplett durchlaufen ist, endet auch die Iteration der verwendenden Schleife. Dann hat die Funktion den impliziten Rückgabewert `None`.
- `return None` stoppt also den Generator
