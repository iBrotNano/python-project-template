- schreibe eine Python-Funktion, welche einen „Zähler“ für eine Liste von Wörtern erstellt und alle Elemente entfernt, deren Anzahl unter einem bestimmten Wert liegt.
- Es bieted sich "Dictionary"-Comprehension an

```python

from collections import Counter

def trim_list(the_list,min_count = 2):

    counter = Counter(the_list)

    
    filtered_counter = {word: count for word, count in counter.items() if count >= min_count}
    
    return filtered_counter # normaler dict



res = trim_list(["Red", "Green", "Black", "Black", "Red", "red", "Orange", "Pink", "Pink", "Red", "White"],min_count = 2)
print(res)
```
