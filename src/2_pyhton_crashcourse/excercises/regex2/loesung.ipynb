Schreiben Sie ein Python-Script text_match, um Folgen von Kleinbuchstaben zu finden, die durch einen Unterstrich verbunden sind.


print(text_match("aab_cbbbc")) # True
print(text_match("aab_Abbbc")) # False
print(text_match("Aaab_abbbc")) # False

```python
import re
def text_match(text):
        patterns = '^[a-z]+_[a-z]+$'
        if re.search(patterns,  text):
                return 'Found a match!'
        else:
                return('Not matched!')

print(text_match("aab_cbbbc"))
print(text_match("aab_Abbbc"))
print(text_match("Aaab_abbbc"))
```
