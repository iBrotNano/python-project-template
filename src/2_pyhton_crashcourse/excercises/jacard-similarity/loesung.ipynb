Schreiben eine Python-Funktion um mithilfe von „Counter“-Objekten den _***Jaccard-Ähnlichkeitskoeffizienten**_ zwischen zwei Listen zu ermitteln.

_Jaccard-Ähnlichkeit_, auch bekannt als _Jaccard-Index_ oder _Jaccard-Koeffizient_, ist ein Maß zur Quantifizierung der Ähnlichkeit zwischen zwei Mengen. Es wird häufig in verschiedenen Bereichen eingesetzt, darunter Data Mining, Informationsabruf und Verarbeitung natürlicher Sprache, um Ähnlichkeiten zwischen Elementsätzen zu vergleichen.

Betrachten Sie beispielsweise zwei Sätze:

A = {Apfel, Banane, Orange, Kiwi}

B = {Banane, Kiwi, Ananas}

Die Schnittmenge von A und B ist {Banane, Kiwi} mit einer Kardinalität von 2. Die Vereinigung von A und B ist {Apfel, Banane, Orange, Kiwi, Ananas} mit einer Kardinalität von 5.  
Dann beträgt die Jaccard-Ähnlichkeit zwischen  A und B = 2/5, also 0,4.

```python
from collections import Counter

list1 = ['Red', 'Green', 'Blue', 'Orange']
list2 = ['Green', 'Pink', 'Blue']

print(set(list1).intersection(set(list2)) )
print(set(list1).union(set(list2)))

print(len(set(list1).intersection(set(list2))))
```

```python

def jaccard_similarity(list1, list2):
    
    jaccard_coefficient = len(set(list1).intersection(set(list2))) / len(set(list1).union(set(list2)))
    return jaccard_coefficient

def main():
    list1 = ['Red', 'Green', 'Blue', 'Orange']
    list2 = ['Green', 'Pink', 'Blue']
    
    jaccard_coefficient = jaccard_similarity(list1, list2)
    print("List 1:", list1)
    print("List 2:", list2)
    print("Jaccard Similarity Coefficient:", jaccard_coefficient)


main()
```

```python

# Variante mit Counter
from collections import Counter

def jaccard_similarity(list1, list2):
    counter1 = Counter(list1)
    counter2 = Counter(list2)
    print(type(counter1))
    intersection_count = sum((counter1 & counter2).values())
    union_count = sum((counter1 | counter2).values())
    
    jaccard_coefficient = intersection_count / union_count
    return jaccard_coefficient

def main():
    list1 = ['Red', 'Green', 'Blue', 'Orange']
    list2 = ['Green', 'Pink', 'Blue']
    
    jaccard_coefficient = jaccard_similarity(list1, list2)
    print("List 1:", list1)
    print("List 2:", list2)
    print("Jaccard Similarity Coefficient:", jaccard_coefficient)


main()
```
