Schreibe eine Python-Funktion um zu überprüfen, ob eine Zeichenfolge nur einen bestimmten Satz von Zeichen enthält (in diesem Fall a-z, A-Z und 0-9).
- Hinweis: [^a-zA-Z0-9] matcht alle Strings die NICHT aus der Menge {a-zA-Z0-9} bestehen.

```python
import re
def only_allowed_chars(string):
    has_unallowed_chars = re.search(r'[^a-zA-Z0-9]',string)
    # man sucht nach Zeichen die nicht a-z,A-Z und 0-9 sind
    return not bool(has_unallowed_chars) # Wenn welche gefunden werden wird das Egebnis negiert um der 
    # Anforderung zu entsprechen (leerer String => False ,d.h. not bool("") => True)

print(only_allowed_chars("ABCDEFabcdef123450")) 
print(only_allowed_chars("*&%@#!}{"))
print(only_allowed_chars("ABC*"))
print(only_allowed_chars("")) # => True hat keine verbotenen Zeichen

```
