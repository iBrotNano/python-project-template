```python
# 1. Einen 10 Pixel breiten Rand um das Bild Zeichnen

from PIL import Image #  "Python Image Library" Pillow Library https://python-pillow.org/
import numpy as np 

im = np.array(Image.open('img/tiger.jpg')) #
print(type(im))
print(im.dtype)
print(im.shape)

#print(im[0,0])
# Rotanteil R, der Grünanteil G und der Blauanteil B die Farbe beschreiben: Farbe = (R, G, B) 
# [255,255,255] = Weiß, [0,0,0] = schwarz

# Bsp ein Weißes Rechteck zeichnen:
im[:10,:] = [255,255,255] 
im[:,-10:]= [255,255,255]
im[-10:,:] = [255,255,255] 
im[:,:10] = [255,255,255] 

# Bild speichern

pil_img = Image.fromarray(im)
print(pil_img.mode)
# RGB
pil_img.save('img/out/tiger.jpg')

# Bild in jupyter-notebook ausgeben
from IPython.display import Image
Image(filename='img/out/tiger.jpg') 
```

```python
# 2. Das Bild jeweils horizontal und vertikal spiegeln

from PIL import Image #  "Python Image Library" Pillow Library https://python-pillow.org/
import numpy as np 

im = np.array(Image.open('img/tiger.jpg')) #
print(type(im))
print(im.dtype)
print(im.shape)

#print(im[0,0])
# Rotanteil R, der Grünanteil G und der Blauanteil B die Farbe beschreiben: Farbe = (R, G, B) 
# [255,255,255] = Weiß, [0,0,0] 

#im = im[::-1,:]
#im = im[::,::-1]
im = im[::-1,::-1] # beides


# Bild speichern

pil_img = Image.fromarray(im)
print(pil_img.mode)
# RGB
pil_img.save('img/out/tiger.jpg')

# Bild in jupyter-notebook ausgeben
from IPython.display import Image
Image(filename='img/out/tiger.jpg') 
```

```python
# 3. Ein zweites Bild einfügen

from PIL import Image #  "Python Image Library" Pillow Library https://python-pillow.org/
import numpy as np 

im = np.array(Image.open('img/tiger.jpg')) #
im2 = np.array(Image.open('img/schmetterling.jpg')) #

im2 = im2[::10,::10]
#print(im2.shape)


xpos = 130
ypos = 40

im[xpos:xpos+im2.shape[0],ypos:ypos+im2.shape[1]] = im2



# Bild speichern
pil_img = Image.fromarray(im)
print(pil_img.mode)
# RGB
pil_img.save('img/out/tiger.jpg')

# Bild in jupyter-notebook ausgeben
from IPython.display import Image
Image(filename='img/out/tiger.jpg') 

```

```python
# 4.  Füge eine Zwei gleiche Bilder aneinander mit der Funktion `concatenate`  
#(<https://numpy.org/doc/stable/reference/generated/numpy.concatenate.html>)

from PIL import Image #  "Python Image Library" Pillow Library https://python-pillow.org/
import numpy as np 

im = np.array(Image.open('img/tiger.jpg')) #


#im = np.concatenate((im,im), axis=1)
im = np.concatenate((im,im), axis=0)

# Bild speichern

pil_img = Image.fromarray(im)
print(pil_img.mode)
# RGB
pil_img.save('img/out/tiger.jpg')

# Bild in jupyter-notebook ausgeben
from IPython.display import Image
Image(filename='img/out/tiger.jpg') 
```

```python
# 5. rotiert das Bild  um 90 Grad und benutze dabei die Funktion numpy.rot90()   
# (<https://numpy.org/doc/stable/reference/generated/numpy.rot90.html>)  

from PIL import Image #  "Python Image Library" Pillow Library https://python-pillow.org/
import numpy as np 

im = np.array(Image.open('img/tiger.jpg')) #
print(type(im))
print(im.dtype)
print(im.shape)

#im = np.rot90(im)
im = np.rot90(im,k=1)
#im = np.rot90(im,k=2,axes= (0,1))
#im = np.rot90(im,k=2,axes= (1,2))


# Bild speichern
pil_img = Image.fromarray(im)
print(pil_img.mode)
# RGB
pil_img.save('img/out/tiger.jpg')

# Bild in jupyter-notebook ausgeben
from IPython.display import Image
Image(filename='img/out/tiger.jpg') 

```
