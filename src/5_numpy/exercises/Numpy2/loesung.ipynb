```python
import numpy as np

A = np.array([[4, 3,-1], [-3, -4, 5],[-2,2,1]])

# Bezeichnung : detA ist die Determinante von A


# Wir brauchen ja detA, detA1,detA2,detA3
# wobei detA1 der Determinante der Matrix A entspricht bei der die 1 Spalte  mit der letzen Spalte (hinter dem =) => (2,-5,6) vertauscht wird
# detA2,detA3 Analog
# als A1 bezeichnen wir die 3x3 Matrix die durch das vertauschen ensteht

# A2,A3 analog
# Die Ergebnisse kann man dann mit detA1/detA, detA2/detA, detA3/detA

# Zunächst müssen wir A1-A3 als Numpy-Array haben

# 1. Möglichkeit: wir schreiben A1-A3 wie A mittels einer Listendarstellung  "manuell"
A1 = np.array([[2, 3,-1], [-5, -4, 5],[6,2,1]])
A2 = np.array([[4, 2,-1], [-3, -5, 5],[-2,6,1]])
A3 = np.array([[4, 3,2], [-3, -4, -5],[-2,2,6]])


# 2. Möglichkeit Slicing
#b = np.array([[2], [-5],[6]]) # Fehler !

b = np.array([2,-5,6])

A1 = np.copy(A)
A1[:, 0] = b

#print(A1)

A2 = np.copy(A)
A2[:, 1] = b

A3 = np.copy(A)
A3[:, 2] = b

#exit()


# => 
detA =  np.linalg.det(A)
detA1 = np.linalg.det(A1)
detA2 = np.linalg.det(A2)
detA3 = np.linalg.det(A3)

detA =  int(round(np.linalg.det(A)))
detA1 = int(round(np.linalg.det(A1)))
detA2 = int(round(np.linalg.det(A2)))
detA3 = int(round(np.linalg.det(A3)))



x =  detA1/detA
y = detA2/detA
z = detA3/detA



print(x,y,z) 

# nicht ganz was wir erwartet haben aber sehr nah dran 
# Warum ist das so ? 
# => https://stackoverflow.com/questions/588004/is-floating-point-math-broken


# 1. Verbesserungs-Idee 
# cast zu int()

# aber was ist mit -1.99999999999

# 2. Verbesserungs-Idee 
# int(round(x))

#detA =  int(round(np.linalg.det(A)))
#detA1 = int(round(np.linalg.det(A1)))
#detA2 = int(round(np.linalg.det(A2)))
#detA3 = int(round(np.linalg.det(A3)))

```
