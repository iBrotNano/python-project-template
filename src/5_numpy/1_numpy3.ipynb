## Performance-Test: Numpy vs Pyhton-Lists

```python
#2
# Ein Array und eine Liste mit jeweils 1000000 Einträgen um damit zu rechnen

my_arr = np.arange(1000000) # ähnlich zu range() 
my_list = list(range(1000000))
my_arr
```

#### Idee: Beide Datentypen werden Elementweise mit Zwei multipliziert
- __np.array:__ `my_arr *2`
- __list:__ `[i*2 for i in my_list]`. 
  
Das Modul `timeit` in Python wird verwendet, um die Ausführungszeit kleiner Codeausschnitte zu messen. Es bietet eine einfache Möglichkeit, die Performance von Code zu analysieren und zu optimieren. 

- diese Art der Berechnung mit Numpy nennt man __vektorisierte Berechnung__. Vektorisierte Berechnung bezeichnet die Anwendung mathematischer Operationen auf alle Elemente eines Vektors (oder Arrays) gleichzeitig, anstatt mit einzelnen Schleifen, um Berechnungen insbesondere bei großen Datenmengen zu beschleunigen.

- Vorteile sind _Performance_ und eine _vereinfachte Schreibweise_.



```python
#3
# Array *2
%timeit -n 100 -r 10 my_arr * 2

#751 mikro-sekunden = 0.000751 Sekunden
```

```python
# List * 2
%timeit -n 100 -r 10 [i*2 for i in my_list]

#29ms = 29 Millisekunden = 0.029
```

##### Magic Befehl: `%timeit -n -r` 
- insgesamt wird der Code n*r mal ausgeführt
- __n loops:__ Um eine einzelne, messbare Dauer pro "Durchlauf" zu erhalten, indem der Code-Ausschnitt viele Male wiederholt wird.
- __r repeats:__ Um externen Systemschwankungen (z. B. CPU-Last, Garbage Collection) Rechnung zu tragen, wird das gesamte Timing-Experiment mehrmals wiederholt.

```python
#19
# Zufällig -- 
# Gleichverteilung
from numpy.random import default_rng
print(np.random.default_rng(42).random((2,3))) # 42 ist das "seed", shape ist 2x3
# oder einfach
print(np.random.random((3,3))) # 3x3


# Normalverteilung (Glockenkurve) 
from numpy.random import normal
mu, sigma = 0, 0.1 #  mu <=> mittelwert, sigma <=> Standard-Abweichung
g = normal(mu, sigma, 500)
print(g)
#vgl.:

```

#### Warum ist das so wichtig?

Die Normalverteilung ist deshalb so hilfreich, weil sie vorhersagbar ist. Mit Hilfe der Standardabweichung ($\sigma$) und dem Mittelwert ($\mu$) kann man genau berechnen, wie wahrscheinlich ein bestimmtes Ereignis ist

Beispiele sind:

- Körpergröße von Erwachsenen
    Mittelwert in Deutschland 179 Männer (Frauen 166) , Standard-Abweichung 7 (Frauen 6.4)

- IQ-Werte (Intelligenzquotient)
    - Der IQ ist normalverteilt, das heißt, sehr hohe oder sehr niedrige Werte sind selten, während Werte nahe 100 am häufigsten vorkommen. Mittelwert ist 100. und die Standard-Abweichung 15 Punkte
