# Array Indexing & Slicing: 

### Indexing und Slicing bei 1 dimensionalen Arrays

```python
import numpy as np

x = np.array([5, 0, 3, 4, 7, 9])
print(type(x))

#indexing
print("x[0]: ",x[0])
print("x[3]: ",x[3])
print("x[-1]: ",x[-1]) # negative indices sind erlaubt !

#slicing
print("x[0:2]: ",x[0:2],type(x)) # teil_array[von:bis_exclusive]
print("x[0:5:2]: ",x[0:5:2],type(x))

```

### Indexing bei Mehrdimensionalen Arrays

- Beim Indexing ist der "End-Index" INCLUSIVE aber NICHT bei SLICING (',') !!
- Beim Indexing von mehrdimensionalen Arrays

```python

np.random.seed(42) # Reproduzierbare Ergebnisse
A = np.random.randint(0, 100, (6, 6))

print(A)

#Indexing:
print("A[0,3]: ",A[0,3])# Index beginnt bei 0, - 3 inklusive
print("A[2,3]: ",A[2,3]) # 3 inklusive => End-Index inklusive !
print("A[2,-1]: ",A[2,-1]) # -1 inklusive

```

##### Warum resultieren die Arrays in unterschiedlichen Shapes?
___Gemini___:  
"Der Hauptunterschied besteht darin, dass beim  _Slicing_ die Dimensionen des Arrays <u>erhalten</u> bleiben, während sie durch "Indexing" <u>reduziert</u> werden können. 


#### Beispiel

```python
#Beispiel: 
A = np.arange(9).reshape((3,3))

A
#A[2] # array([6, 7, 8]) 
#A[2].shape
#A[2].ndim

#b)

A[2:,:] # array([[6, 7, 8]])
#A[2:,:].shape # (1, 3)
#A[2:,:].ndim


B = A[2:,:].reshape((3,))
B
#B.shape

```

#### Was ist der Unterschied? 

__(3,) Shape: 1-Dimensional Array__

Ein Array mit der Shape (3,) ist ein __1-dimensionales array__, auch Spalten-Vektor genannt. Es hat eine Axe 0 mit der Länge 3. 
Man kann es sich als Liste von  Zahlen vorstellen.

__(1, 3) Shape: 2-Dimensional Array__

Ein Array der Shape (1, 3) ist ein __2-dimensional array__, auch Matrix genannt. Es hat zwei Achsen:

- Axis 0 (rows): der Länge 1.

- Axis 1 (columns): der Länge 3.

Man kannn es sich als Tabelle mit einer Zeile und drei Spalten vorstellen.

___Zusammenfassend lässt sich sagen, dass ein (3,) Array ein eindimensionaler Vektor ist, während ein (1,3) Array eine zweidimensionale Matrix ist.___

#### Ist dieser Unterschied wichtig?

In manchen Situationen: 

- manche Framworks/ Algorithmen brauchen eine bestimmte Shape. (bsp.: `scikit-learn` )


- Es macht einen Unterschied bei der Matrizenmultiplikation (`np.dot` oder `@`):

    - Die Multiplikation eines (1,3)-Arrays mit einem (3,1)-Array ist möglich und ergibt eine Matrix der Shape (1,1).

    - Die Multiplikation eines (3,)-Arrays mit einem (3,)-Array ist als Matrizenmultiplikation nicht definiert, da die Dimensionen nicht kompatibel sind. Stattdessen führt np.dot(v, v) eine Punktprodukt-Operation durch und gibt einen Skalar der Shape "()" zurück.

### Array Manipulation

```python
# Daten können auf diese Weise auch manipuliert werden:
data = np.array([(4,7),(0,2),(-5,6),(0,0)])

print(data)
print("----")

data[data < 0] = 0 # 
print(data)

print("----")
print(data[names != 'joe'])
print("----")
data[names != 'joe'] = 7
print(data)
```

### Achtung: 
Beim Slicing mit NumPy werden im Gegensatz zu Standard-Python-Slicing <u>keine Kopien erzeugt</u>, stattdessen spricht man von __Views__

- Basic Slicing/Indexing erzeugt Views 


```python
arr = np.arange(9).reshape((3,3))

print(arr)
print("----")
view = arr[:2,:]
print(view)
view[0,0] = -1
print(arr) # (!)

print("------")
# Möchte man eine Kopie erzeugen kann man .copy() verwenden
arr = np.arange(9).reshape((3,3))
my_copy = arr[:2,:].copy()
print(my_copy)
my_copy[0,0] = -2
print(arr)


```

#### Basic Indexing erzeugt einen View

```python
# Indexing gerzeugt eine Kopie
print("-- Indexing:")
arr = np.arange(9).reshape((3,3))
print(arr)
no_view = arr[1]
no_view[0] = -1
print("----")
print(no_view)
print("----")
print(arr)
```

### Advanced Indexing erzeugt eine Kopie

- "Fancy"-Indexing
- Boolean-Indexing

```python
print("-- Fancy-Indexing:")
arr = np.arange(9).reshape((3,3))
print(arr)
print("------")
no_view = arr[[0,1,2]]
no_view[0,0] = -1
print(no_view)
print("------")
print(arr)

```

```python
print("-- Boolean-Indexing:")
arr = np.arange(9).reshape((3,3))
print(arr)
tmp = arr[(arr > 4)]
tmp[0] = 100
print(tmp)
print(arr)

```
