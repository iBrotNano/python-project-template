# Numpy Universal-Functions (UFunctions):

* __Numpy-Universal-Funktions__ sind _Vektorisierte-Operationen_ auf Numpy-Arrays 
* __Vektorisierte-Operationen__ sind schneller als die _Iteration_ über Arrays. 
* Der _Vektorisierte Ansatz_ sorgt dafür, daß die Iterationen in die "kompilierte Schicht" verlagert werden.  
* Der komplierte Code ist um ein vielfaches schneller als eine klassische Python-Implementierung mit Schleifen, da insbesondere die Typüberprüfung und Funktionsaufrufe wegfallen.  (vgl. HDS Seite 71)



````python
import numpy as np
# Vektorisierte Operationen auf Arrays
#=>
arr = np.arange(9).reshape((3,3))
arr = arr + 5
arr
#<=


````

Wie schnell ist diese Operation ?

````python
#%timeit
arr = np.arange(1000).reshape((10,100))
%timeit -n 1000 -r 100 arr + 5
# 977 ns ± 418 ns per loop (mean ± std. dev. of 100 runs, 1,000 loops each)
````

````python
#Re-Creation
arr = np.arange(1000).reshape((10,100))
````

````python

%%timeit -n 1000 -r 100 # muss als ersten in einer Zelle stehen
# selbe Operation mittels Iteration 
#=>
for i in range(np.size(arr,axis=0)):
    for j in range(np.size(arr,axis=1)):
        arr[i,j] = arr[i,j] +5
arr
#<=
#171 μs ± 5.66 μs per loop (mean ± std. dev. of 100 runs, 1,000 loops each)
````

### Vergleich:

* nano (n) 10−9 => 1ns == 0.000000001s <=> ein Milliardstel
* mikro (μ) 10−6 => 1μs == 0.000001s <=> ein Millionstel

* => 977 ns == 0.977 μs <  171 μs =>

## Rechenoperationen auf Arrays mittels Universal-Functions (UFunctions)

* Mit UFunctions lassen sich Rechenoperationen auf Arrays anwenden
    * Arithmetische Operatoren_
        * `+,-,*,/,%,**,//, ..` ; welche auf entsprechenden Numpy-Funktionen basieren:  "+" <=> np.add(), ..
    * Logische Operatoren 
        * `np.logical_and`
        * `np.logical_or`
        * `np.logical_not`
    * Trigonometrische Funktionen: `np.sin, np.cos, np.tan, ..`
    * Exponentiele und Logarithmische Funktionen
    * uvm. vgl.: https://numpy.org/doc/2.1/reference/routines.math.html

* Man unterscheidet _unäre_ und _binäre_ Operation, - also ob bei einer _Operation_ "ein" oder "zwei" _Operanden_ verwendete werden 

* Bei gleicher _Shape_ erfolgen die Operationen ***Elementweise*** (!)

* Unterscheidet sich die Shape der Operanden, dann kommt das sogenannte ___Broadcasting___ ins Spiel.

#### Beispiele : gleiche Shape, Elementweise Addition

````python
# Beispiel 1:
import numpy as np

A = np.arange(9).reshape((3,3))
B = np.arange(9).reshape((3,3))
print(A)
print(B)
A+B

````

````python
# Beispiel 2:
x = np.arange(4).reshape((2,2)) #
y = np.arange(4).reshape((2,2)) #

print(x)
print(y)
print(x+y)
print(x*y) # 
````

#### Bei unterschiedlicher Shape kommt __"Broadcasting"__ ins Spiel
Beim Broadcasting werden die Arrays erweitert, so daß die Shape zueinander passt.

<img src="img/broadcasting.png" stlye="width:400px;">


### Broadcasting Regeln:

##### Regel 1
* Wenn sich die Anzahl der Dimensionen der beiden Arrays unterscheidet, wird die Shape des Arrays mit der niedrigeren Dimension auf der vorderen (linken) Seite mit Einsen aufgefüllt.
##### Regel 2
* Wenn die Shapes beider Arrays in irgendeiner Dimension nicht übereinstimmen, wird das Array, dessen Shape in dieser Dimension gleich 1 ist, ausgedehnt, um zu der anderen Shape zu passen.
##### Regel 3
* Wenn in irgendeiner Dimension die Größen der beiden Arrays nicht übereinstimmen und keine davon gleich 1 ist, kommt es zu einer Fehlermeldung. - Also ist Broadcasting nicht möglich


#### Beispiel 1: 

```python
A = np.arange(4)
B = 5

# shape (4,) und (1,) 
# Regel 2 => B wird ausgedehnt so das es Shape (4,) hat
# A <=> [0,1,2,3] und B <=> [5,5,5,5]
```

````python
# Beispiel 1
import numpy as np

x = np.arange(4) # 1D Array
y = 5

print("x =", x)
print("x + 5 =", x + y) # np.add(x,5)
print("x - 5 =", x - y) # np.sub(x,5)
print("x * 2 =", x * y) #
print("x / 2 =", x / y)
print("x // 2 =", x // y)  # floor division 

````

#### Aggregationsfunktionen

* Aggregationsfunktionen sind Funktionen welche als Eingabe eine Menge von Werten haben und als Ergebnis einen einzigen Wert haben
* z.B. das größte Element einer Zahlenmenge

````python
# paar Beispiele:
L = np.random.random(1000)
#print(L)
print("-----")
print(np.sum(L))
print(np.max(L))
print(np.min(L))
print(np.mean(L))
````

````python
# Bei Mehrdimensionalen Arrays kann die dimension/achse angegben werden
M = np.random.random((3, 4))
print(M)
print("---")
print(M.min(axis=0))# Entlang der Achse 0 
````

# Boolsche Operatoren und Arrays

````python
a = np.array([[True, False, True],[True, False, True],[True, False, True]])
b = np.array([[True, True, True],[True, False, True],[False, False, True]])
print(a,b,sep="\n-----\n")

print("\n----------------------\n")

c = np.logical_and(a,b)
d = np.logical_or(a,b)
e = np.logical_not(a)

print(c,d,e,sep="\n------\n")




````

* np.any() testet ob EINS der Elemente True ist
* np.all() testet ob ALLE Elemente True ist

````python
# Bsp.:
a = np.array([[True, False, True],[True, False, True],[True, False, True]])
b = np.array([[True, True, True],[True, False, True],[False, False, True]])
c = np.array([[True, True, True]])



print(np.any(a)) # True es gibt ein True Element
print(np.all(b)) # False - nicht alle Elemente sind True
print(np.all(c)) # True alle Elemente sind True


````
