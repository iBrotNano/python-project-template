## Numpy-Arrays in eine Datei schreiben und lesen

```python

import numpy as np

#Array in Datei schreiben
arr = np.arange(20).reshape((4,5))

np.save('data/myarray.npy',arr) 

```

```python
#Array in Datei lesen
arr_loaded = np.load('data/myarray.npy')
print(arr_loaded)
```

```python

# Array Archiv (uncompressed) ".npz"
arr1 = np.arange(20).reshape((4,5))
arr2 = np.arange(20).reshape((2,5,2))

np.savez('data/array_archive.npz', first=arr, second=arr2, third=arr1) # **kwargs

arch = np.load('data/array_archive.npz')
print("---")
print('second:\n',arch['second'])
print('third:\n',arch['third'])
```
