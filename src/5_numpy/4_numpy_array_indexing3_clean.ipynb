# Array Indexing & Slicing: 

### Indexing und Slicing bei 1 dimensionalen Arrays

```python
import numpy as np

x = np.array([1,2,3,4,5,6])
type(x)

#indexing
print(x[0])
print(x[-1]) # negative indices sind erlaubt

#slicing
print(x[0:2]) # [1 2]
print(x[0:5:2],type(x[0:5:2]))


x = np.array([[1,2,3],[4,5,6]])
print(x[0][0])
```

### Indexing bei Mehrdimensionalen Arrays

- Beim Indexing ist der "End-Index" INCLUSIVE aber NICHT bei SLICING (',') !!
- Beim Indexing von mehrdimensionalen Arrays

```python
np.random.seed(42)
A = np.random.randint(0,100,(6,6))
print(A)

#Indexing
print(A[2,2]) # Index beginnt mit 0, 2 inclusive
print(A[-1,-1]) 

```

#### Was ist der Unterschied? 

__(3,) Shape: 1-Dimensional Array__

Ein Array mit der Shape (3,) ist ein __1-dimensionales array__, auch Spalten-Vektor genannt. Es hat eine Axe 0 mit der Länge 3. 
Man kann es sich als Liste von  Zahlen vorstellen.

__(1, 3) Shape: 2-Dimensional Array__

Ein Array der Shape (1, 3) ist ein __2-dimensional array__, auch Matrix genannt. Es hat zwei Achsen:

- Axis 0 (rows): der Länge 1.

- Axis 1 (columns): der Länge 3.

Man kannn es sich als Tabelle mit einer Zeile und drei Spalten vorstellen.

___Zusammenfassend lässt sich sagen, dass ein (3,) Array ein eindimensionaler Vektor ist, während ein (1,3) Array eine zweidimensionale Matrix ist.___

#### Ist dieser Unterschied wichtig?

In manchen Situationen: 

- manche Framworks/ Algorithmen brauchen eine bestimmte Shape. (bsp.: `scikit-learn` )


- Es macht einen Unterschied bei der Matrizenmultiplikation (`np.dot` oder `@`):

    - Die Multiplikation eines (1,3)-Arrays mit einem (3,1)-Array ist möglich und ergibt eine Matrix der Shape (1,1).

    - Die Multiplikation eines (3,)-Arrays mit einem (3,)-Array ist als Matrizenmultiplikation nicht definiert, da die Dimensionen nicht kompatibel sind. Stattdessen führt np.dot(v, v) eine Punktprodukt-Operation durch und gibt einen Skalar der Shape "()" zurück.
