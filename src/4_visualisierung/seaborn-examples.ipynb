#### Seaborn erzeugt "attraktive" Diagramme

```python
import matplotlib.pyplot as plt
import seaborn
import numpy as np
#seaborn.set()  # Deprecated (!) stattdessen plot style
#seaborn.set_theme()

plt.style.use("seaborn-v0_8-whitegrid")
#plt.style.use("seaborn-v0_8-deep")
#plt.style.use("seaborn-v0_8-pastel")

# Seaborn zurücksetzen
#seaborn.reset_defaults()


ywerte = np.random.randint(0,100,7)
xwerte = ["spd", "cdu", "afd", "grüne", "linke", "Frei Wähler", "Basis"]

plt.bar(xwerte, ywerte)
plt.xlabel("X-Werte")
plt.ylabel("Y-Werte")
plt.show()
```

```python
# Import seaborn
import seaborn as sns
penguins = sns.load_dataset("penguins") # Pandas Dataframe
display(penguins.head(3))
#penguins.head(4)

#hue: Semantic variable that is mapped to determine the color of plot elements.

sns.jointplot(data=penguins, x="flipper_length_mm", y="bill_length_mm", hue="species")

```
