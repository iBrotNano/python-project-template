```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#Seaborn Style verwenden
#import seaborn 
#seaborn.set() # Einschalten
#seaborn.reset_defaults() # Ausschalten

data = pd.read_csv('csv/president_heights.csv' )
#print(type(data))

#data.head(10)
#data.tail(10)

heights = data['height(cm)']

#print(type(heights))

plt.hist(heights,bins=10)

plt.title('Height Distribution of US Presidents')
plt.xlabel('Height (cm)')
plt.ylabel('Number')

plt.show()
```
