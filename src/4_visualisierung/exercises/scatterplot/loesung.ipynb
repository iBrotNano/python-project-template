Schreibe ein Python-Script, um ein Scatterplot (Streudiagramm) zu zeichnen, das eine zufällige Verteilung in X und Y annimmt und gegeneinander aufträgt.
- verwende z.B.  
`random.random()`  

Beispiel Ausgabe:

![](img/scatterplot.png)

```python
import matplotlib.pyplot as plt
import random

X = [ random.random() *10 for _ in range(50)]
Y = [ random.random() *10 for _ in range(50)]
plt.scatter(X,Y, color='r')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
