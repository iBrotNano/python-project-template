- Die Sigmoid Funktion wird z.B als Aktivierungsfunktion in Neuronalen Netzen verwendet  
https://en.wikipedia.org/wiki/Sigmoid_function
- Zeichne Sie mit Matplotlib

```python
import math
import matplotlib.pyplot as plt

# definition der (x,y) Wertpaare
x = [x / 10.0 for x in range(-50, 50, 1)]
y = [ 1/(1+math.exp(-s)) for s in x]

# schreiben der (x,y)-Wertpaare 
plt.plot(x, y)

#Legende
plt.title("Sigmoid", loc="left")
plt.xlabel("x-Achse")
plt.ylabel("y-Achse")

#Layout
plt.margins(0.2, 0.2) # nicht unbedingt nötig

plt.show()
```
