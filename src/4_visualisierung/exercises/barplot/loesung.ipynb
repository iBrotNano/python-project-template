### Barplots
  
- Erstelle einen Bar-Chart der alle US-Präsidenten aus der csv-datei mit IQ (_IQ_) darstellt
- überlege dir ggf. weitere Datenfelder die für dich interessant sind und stelle diese mit Matplotlib dar
- Hinweis: Die Daten müssen ggf. angepasst werden um sie verwenden zu können


```python
import matplotlib.pyplot as plt 
import csv 
import re

x = []
y = []

with open('US-Presidents.csv','r') as csvfile:
    plots = csv.reader(csvfile,delimiter=',')
    next(plots,None) # Skip Header
    for row in plots:
        x.append(row[1]) # Name
        tmp = re.split("-",row[11]) # iq: 120-160
        if(len(tmp) == 2):    
            #y.append(float( tmp[1])) # IQ
            y.append(float( tmp[0])) # Soll hier der höhere oder niedrigere IQ verwendete werden ?
        else:
            y.append(0 if tmp[0]== '' else float(tmp[0]))


plt.bar(x,y,label='Height', color='b')
plt.xticks(rotation='vertical',fontsize=8)

plt.xlabel('Presidents') 
plt.ylabel('IQ') 
plt.title('US-Presidents IQs') 
plt.legend() 
plt.show() 
```
