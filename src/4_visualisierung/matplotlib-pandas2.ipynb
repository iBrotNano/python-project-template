```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#Seaborn Style verwenden
#import seaborn 
#seaborn.set() # Einschalten
#seaborn.reset_defaults() # Ausschalten


data = pd.read_csv('csv/president_heights.csv')

#display(data)
#print(type(data))

heights = data['height(cm)']

plt.hist(heights,bins=15)

plt.title('Height Distribution of US Presidents')
plt.xlabel('height (cm)')
plt.ylabel('number')

plt.show()
```
